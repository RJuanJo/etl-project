## Transformations

- What are the columns that serve us? In this dataset we see excesses of columns that do not provide us with in-depth information for our analysis, so we will look for which ones are not useful to us and we will carry out a deterioration to do a more detailed analysis.


```python
column_names_json = dataU.columns.to_list()
column_names_json = pd.Series(column_names_json).to_json(orient='values')
column_names_json
```

```python
drop_columns = ["url","photos","reviews","photos","isAvailable","additionalHosts","primaryHost","isHostedBySuperhost"]
dataU = dataU.drop(columns=drop_columns)
```

```python
column_names_json = dataU.columns.to_list()
column_names_json
```

```python
def calification(stars):
    if 0.0 <= stars <= 2.0:
        return 'Acceptable'
    elif 2.1 <= stars <= 3.5:
        return 'Good'
    elif 3.6 <= stars <= 4.0:
        return 'Very good'
    elif 4.1 <= stars <= 5.0:
        return 'Excellent'
    else:
        return 'Unknown' 

dataU['calification'] = dataU['stars'].apply(calification)
print(dataU[['stars', 'calification']].head(20))
```
